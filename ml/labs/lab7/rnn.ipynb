```python
import numpy as np
import keras

from collections import defaultdict

from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

from modules.textProcessor import read_text, get_char_indexes, get_data_from_text, get_coding

np.random.seed(42)
```

```python
text = read_text('./data/text.txt')
chars, indexes = get_char_indexes(text)
data_x, data_y = get_data_from_text(text)
x, y = get_coding(data_x, data_y, chars)

print(text)
print(x.shape)
print(y.shape)
```

```python
def get_model_lstm():
    model = Sequential()

    model.add(LSTM(256,  input_shape=(x.shape[1], x.shape[2]), return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(100))

    model.add(Dense(y.shape[1], activation='softmax'))

    model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(), metrics=['accuracy'])

    return model
```

```python
model = get_model_lstm()
```

```python
model.fit(x, y, batch_size = 64, epochs=64)
```

```python
for _ in range(5):
    random_index = np.random.randint(len(data_x) - 1)
    current_pattern = data_x[random_index]
    ans = current_pattern

    print(f'Pattern is {current_pattern}')

    for _ in range(30):
        x = np.zeros((1, 5, len(sorted(list(set(text))))))
        for i, char in enumerate(current_pattern):
            x[0, i, chars[char]] = 1
        predict = model.predict(x)
        next = indexes[np.argmax(predict)]
        current_pattern = current_pattern[1:] + next
        ans += next

    print(f'Predict is {ans}')
    print('-----------------')
```

```python
continues = defaultdict(lambda: defaultdict(lambda: 0))
for sentence, continuee in zip(data_x, data_y):
    continues[sentence][continuee] += 1
print(continues)
```

```python
for _ in range(5):
    random_index = np.random.randint(len(data_x) - 1)
    current_pattern = data_x[random_index]
    ans = current_pattern
    print(f'Pattern is {current_pattern}')
    for _ in range(30):
        nexts = continues[current_pattern]
        nexts_values = [v for v in nexts.values()]

        if len(nexts_values) == 0:
            continue

        next = list(nexts.items())[np.argmax(nexts_values)]
        current_pattern = current_pattern[1:] + next[0]
        ans += next[0]
    print(f'Predict is {ans}')
    print('-----------------')
```

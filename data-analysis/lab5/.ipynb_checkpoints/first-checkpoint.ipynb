```python
import warnings
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import umap.umap_ as umap

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import euclidean_distances

from modules.extractors import salary, height, weight, age

from IPython.core.pylabtools import figsize

warnings.filterwarnings('ignore')
```

```python
SIZE_X = 14
SIZE_Y = 7
figsize(SIZE_X, SIZE_Y)
```

```python
df = pd.read_csv("data/raw/nba2k-full.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(["full_name", "jersey", "team", "country", "college", "version", "draft_year", "draft_round", "draft_peak"], axis=1, inplace=True)
df.salary = df.salary.pipe(salary)
df.height = df.height.pipe(height)
df.weight = df.weight.pipe(weight)
df.rating = df.rating.map(int)
df = df.assign(age=df.b_day.pipe(age))
df.drop("b_day", axis=1, inplace=True)
df.head()
```

```python
print(f"Max salary: {df.salary.max()}")
print(f"Max rating: {df.rating.max()}")
print(f"Max height: {df.height.max()}")
print(f"Max weight: {df.weight.max()}")
print(f"Max age: {df.age.max()}")
```

```python
#df = df[df.position.isin(["C", "F", "G"])]
```

```python
X_parameters = ["height", 
                "weight", 
                "age", 
                "rating", 
                #"salary"
               ]
Y_parameter = "position"
X = df[X_parameters]
Y = df[Y_parameter]
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_PCA_reduced = pca.fit_transform(X)
sns.scatterplot(x=X_PCA_reduced[:, 0], y=X_PCA_reduced[:, 1], hue=Y)
```

```python
df.salary /= 40_000_000
df.rating /= 100
df.height /= 2
df.age /= 40
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_PCA_reduced = pca.fit_transform(X)
sns.scatterplot(x=X_PCA_reduced[:, 0], y=X_PCA_reduced[:, 1], hue=Y)
```

```python
tsne = TSNE(n_components=2, random_state=42, learning_rate='auto')
X_TSNE_reduced = tsne.fit_transform(X)
sns.scatterplot(x=X_TSNE_reduced[:, 0], y=X_TSNE_reduced[:, 1], hue=Y)
```

```python
reducer = umap.UMAP(n_neighbors=10, min_dist=0.3, random_state=42)
X_UMAP_reduced = reducer.fit_transform(X)
sns.scatterplot(x=X_UMAP_reduced[:, 0], y=X_UMAP_reduced[:, 1], hue=Y)
```

```python
#count_of_clasters = 3
count_of_clasters = 7

for X_reduced in [X_PCA_reduced, X_TSNE_reduced, X_UMAP_reduced]:
    kmeans = KMeans(n_clusters=count_of_clasters).fit(X_reduced)
    sns.scatterplot(x=X_reduced[:, 0],
                    y=X_reduced[:, 1],
                    hue=kmeans.labels_,
                    palette=sns.color_palette(n_colors=count_of_clasters))
    plt.show()
    mean_center_distance = np.triu(
        euclidean_distances(kmeans.cluster_centers_), 1).mean()
    print(f"Mean distance between cluster centers: {mean_center_distance:.3f}")
    mean_distance_to_center = np.mean([
        euclidean_distances([kmeans.cluster_centers_[kmeans.labels_[idx]]],
                            [point]) for idx, point in enumerate(X_reduced)
    ])
    print(f"Mean distance to center: {mean_distance_to_center:.3f}")
```
